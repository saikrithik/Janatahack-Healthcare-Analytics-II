```python
df_train = pd.read_csv("/content/drive/My Drive/Colab Notebooks/JanathaHack/Janatahack Healthcare Analytics II/train.csv")
df_test = pd.read_csv("/content/drive/My Drive/Colab Notebooks/JanathaHack/Janatahack Healthcare Analytics II/test.csv")
sub = pd.read_csv("/content/drive/My Drive/Colab Notebooks/JanathaHack/Janatahack Healthcare Analytics II/sample_submission.csv")
df_train = df_train.fillna(-99)
df_test = df_test.fillna(-99)
str_cols = [ col  for col, dt in df_test.dtypes.items() if dt == object]

le2 = LabelEncoder()
df_train["Stay"] = le2.fit_transform(df_train["Stay"])

full_df = pd.concat([df_train,df_test])
full_df = full_df.reset_index(drop = True)

print("LabelEncoding...")
le = LabelEncoder()
full_df["Hospital_type_code"] = le.fit_transform(full_df["Hospital_type_code"])
full_df["Hospital_region_code"] = le.fit_transform(full_df["Hospital_region_code"])
map = {'Emergency': 2, 'Trauma': 3, 'Urgent': 1}
full_df["Type of Admission"] = full_df["Type of Admission"].map(map)
map = {'Extreme': 3, 'Minor': 1, 'Moderate': 2}
full_df["Severity of Illness"] = full_df["Severity of Illness"].map(map)
map ={'0-10': 10,'11-20': 20, '21-30': 30, '31-40': 40, '41-50': 50, '51-60': 60,
      '61-70': 70,'71-80': 80,'81-90': 90,'91-100': 100}
full_df["Age"] = full_df["Age"].map(map)
full_df["Ward_Type"] = le.fit_transform(full_df["Ward_Type"])
full_df["Department"] = le.fit_transform(full_df["Department"])
full_df["Ward_Facility_Code"] = le.fit_transform(full_df["Ward_Facility_Code"])
cat_cols = ["Hospital_type_code","Hospital_region_code","Age",
           "Ward_Type","Department","Ward_Facility_Code","Type of Admission",
            "Severity of Illness"]
#------------------------------------
print("Concat Features")
full_df["CONCAT Hosp_details"] = full_df["Hospital_code"].astype(str) + "_" + full_df["Hospital_type_code"].astype(str) + "_" + full_df["City_Code_Hospital"].astype(str) + "_" + full_df["Hospital_region_code"].astype(str)
full_df["CONCAT Hosp_details"] = le.fit_transform(full_df["CONCAT Hosp_details"])

full_df["CONCAT Hosp_Rooms"] = full_df["Hospital_code"].astype(str) +"_"+ full_df["Available Extra Rooms in Hospital"].astype(str)
full_df["CONCAT Hosp_Rooms"] = le.fit_transform(full_df["CONCAT Hosp_Rooms"])

full_df["CONCAT Hosp_Ward_details"] = full_df["Hospital_code"].astype(str) +"_"+ full_df["Ward_Type"].astype(str) +"_"+ full_df["Ward_Facility_Code"].astype(str)
full_df["CONCAT Hosp_Ward_details"] = le.fit_transform(full_df["CONCAT Hosp_Ward_details"])

full_df["CONCAT Hosp_BedGrade"]  = full_df["Hospital_code"].astype(str) +"_"+ full_df["Bed Grade"].astype(str)
full_df["CONCAT Hosp_BedGrade"] = le.fit_transform(full_df["CONCAT Hosp_BedGrade"])

#------------------------------------
print("Groupby patientid...")
grpby = full_df.groupby(['patientid']).agg({'Admission_Deposit': ['mean', 'max', 'min','sum'],
                                            'Severity of Illness': ['mean', 'min','count'],
                                            "Age" : ['min',"max",'mean'],
                                            "Visitors with Patient" : ['mean', 'max', 'min'],
                                           "City_Code_Patient" : ['mean', 'max', 'min',"count"],
                                            "Type of Admission" : ["mean",'max','min'],
                                            "Hospital_code" : ["nunique"]
                                           })
grpby.columns = ['_patientid_' + '_'.join(c).strip('_') for c in grpby.columns]
full_df = pd.merge(full_df, grpby, on = ['patientid'], how='left')
#------------------------------------
print("groupby nuniques.....")
temp =full_df.groupby(["patientid"]).size().to_frame()
temp.columns= ["patientid_freq"]
full_df = pd.merge(full_df, temp, on = ['patientid'], how='left')

temp =full_df.groupby(["patientid","Hospital_code"]).size().to_frame()
temp.columns= ["patientid_HospCode_freq"]
full_df = pd.merge(full_df, temp, on = ['patientid',"Hospital_code"], how='left')

full_df["patientid_Hospital_codeNUNIQUE"] = full_df.groupby("patientid")["Hospital_code"].transform('nunique')
full_df["Hospital_patientid_codeNUNIQUE"] = full_df.groupby("Hospital_code")["patientid"].transform('nunique')
#------------------------------------
print("Groupby patientid_Visitors...")
grpby = full_df.groupby(['patientid',"Visitors with Patient"]).agg({'Admission_Deposit': ['mean', 'max', 'min','sum'],
                                            'Severity of Illness': ['mean', 'min','max'],
                                            "Age" : ['min','max','mean'],
                                           "City_Code_Patient" : ['mean', 'max', 'min',"count"],
                                            "Type of Admission" : ["mean",'max','min']
                                           })
grpby.columns = ['_patientid_Visitors_' + '_'.join(c).strip('_') for c in grpby.columns]
full_df = pd.merge(full_df, grpby, on = ['patientid',"Visitors with Patient"], how='left')
#------------------------------------
print("Groupby 'Ward_Type','Ward_Facility_Code',Bed Grade...")
grpby = full_df.groupby(['Ward_Type','Ward_Facility_Code',"Bed Grade"]).agg({'Admission_Deposit': ['mean', 'max', 'min','sum'],
                                            'Severity of Illness': ['mean', 'max', 'min'],
                                            "Age" : ['min','mean'],
                                            "Visitors with Patient" : ['mean','max','min'],
                                           "City_Code_Patient" : ['mean','max','min'],
                                            "Type of Admission" : ["mean","max"]
                                           })
grpby.columns = ['_WardBed_' + '_'.join(c).strip('_') for c in grpby.columns]
full_df = pd.merge(full_df, grpby, on = ['Ward_Type','Ward_Facility_Code',"Bed Grade"], how='left')
#------------------------------------
# print("Prev Target Encoding...")

# for i in range(1,10):
    # full_df[f'prev_target_{i}'] = full_df.sort_values(by='patientid')['Stay'].fillna(method='ffill').shift(-i).sort_index()
#------------------------------------
# grpby = full_df.groupby(['patientid','Hospital_code']).agg({'Ward_Type':lambda x: pd.Series.mode(x)[0],
#                                                             #  'Severity of Illness' :lambda x: pd.Series.mode(x)[0],
#                                                             "Type of Admission":lambda x: pd.Series.mode(x)[0]})
# grpby.columns = ['_patientidHospCode_' + '_'.join(c) for c in grpby.columns]
# full_df = pd.merge(full_df, grpby, on = ['patientid','Hospital_code'], how='left')
#------------------------------------
print("Group By Age Ward_type...")
grpby = full_df.groupby(['Age', 'Ward_Type']).agg({'Admission_Deposit': ['mean', 'max', 'min','sum'],
                                                    'Severity of Illness': ['mean','count'],
                                                                          })
grpby.columns = ['_Age_Ward_Type_' + '_'.join(c).strip('_') for c in grpby.columns]
full_df = pd.merge(full_df, grpby, on = ['Age', 'Ward_Type'], how='left')
#---------------------------------Mean Encoding------------------------------
print("mean encoding")
x = full_df[['Hospital_code','Age','CONCAT Hosp_Rooms','Admission_Deposit']]
cols = ['Hospital_code','Age','CONCAT Hosp_Rooms']

kf = sklearn.model_selection.KFold(n_splits = 10, shuffle = False)
for i in cols:    
    x['Mean_Encoded_on'] = np.nan

    for tr_ind, val_ind in kf.split(x):
        X_tr, X_val = x.iloc[tr_ind], x.iloc[val_ind] ## train-test hold out
        x.loc[x.index[val_ind], 'Mean_Encoded_on'] = X_val[i].map(X_tr.groupby(i).Admission_Deposit.mean())

    x = x.rename(index=str, columns={"Mean_Encoded_on": i +"_K_Encoded"})
reqcols  = list(set(cols) - set(x.columns))
full_df[reqcols] = x[reqcols]
#-----------------------------------------------------------------------------
print("PatientId Time...")
num = 1

full_df["patientid_time"] = 1
for i in range(len(full_df)-1):
  if full_df["patientid"][i] == full_df["patientid"][i+1]:
    full_df["patientid_time"][i] = num
    num+=1
  else:
    num = 1
#------------------------------------
df_test = full_df[full_df["Stay"].isnull()].reset_index(drop=True)
df_test.drop("Stay",axis=1,inplace=True)
df_train = full_df[full_df["Stay"].notnull()]
# ------------------------------------


```

```python
from collections import Counter, defaultdict
import random
def stratified_group_k_fold(X, y, groups, k, seed=None):
    labels_num = np.max(y) + 1
    y_counts_per_group = defaultdict(lambda: np.zeros(labels_num))
    y_distr = Counter()
    for label, g in zip(y, groups):
        y_counts_per_group[g][label] += 1
        y_distr[label] += 1

    y_counts_per_fold = defaultdict(lambda: np.zeros(labels_num))
    groups_per_fold = defaultdict(set)

    def eval_y_counts_per_fold(y_counts, fold):
        y_counts_per_fold[fold] += y_counts
        std_per_label = []
        for label in range(labels_num):
            label_std = np.std([y_counts_per_fold[i][label] / y_distr[label] for i in range(k)])
            std_per_label.append(label_std)
        y_counts_per_fold[fold] -= y_counts
        return np.mean(std_per_label)
    
    groups_and_y_counts = list(y_counts_per_group.items())
    random.Random(seed).shuffle(groups_and_y_counts)

    for g, y_counts in sorted(groups_and_y_counts, key=lambda x: -np.std(x[1])):
        best_fold = None
        min_eval = None
        for i in range(k):
            fold_eval = eval_y_counts_per_fold(y_counts, i)
            if min_eval is None or fold_eval < min_eval:
                min_eval = fold_eval
                best_fold = i
        y_counts_per_fold[best_fold] += y_counts
        groups_per_fold[best_fold].add(g)

    all_groups = set(groups)
    for i in range(k):
        train_groups = all_groups - groups_per_fold[i]
        test_groups = groups_per_fold[i]

        train_indices = [i for i, g in enumerate(groups) if g in train_groups]
        test_indices = [i for i, g in enumerate(groups) if g in test_groups]

        yield train_indices, test_indices
```
